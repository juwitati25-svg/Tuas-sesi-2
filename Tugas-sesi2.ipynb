```python
#1. Spilt Bill Nongkrong
def split_bill():
  print("\n=== Spilt Bill Nongkrong ===")
  total_tagihan = float(input("Masukan total tagihan:"))
  jumlah_orang = int(input("Masukan jumlah orang:"))
  pajak = 0.10

  total_setelah_pajak = total_tagihan + (total_tagihan * pajak)
  per_orang = total_setelah_pajak / jumlah_orang

  print("Total stelah pajak:", total_setelah_pajak)
  print("Bayar per orang:", per_orang)

split_bill()
```

```python
#2. Konvesi Waktu Kerja Freelance
def freelance():
  print("\=== Konversi Waktu Freelance ===")
  jam = int(input("Masukan jam kerja:"))
  menit = int(input("Masukan menit kerja:"))
  upah_per_jam = float(input("Masukan upah per jam:"))

  total_jam = jam + (menit / 60)
  total_bayar = total_jam * upah_per_jam

  print("Total jam (desimal):", total_jam)
  print("Total bayaran:", total_bayar)

```

```python
from ast import keyword
#3. Estimasi BBM Perjalanan
def estimasi_bbm():
  print("\n=== Estimasi BBM ===")
  jarak = float(input("Masukan jarak perjalanan (km):"))
  konsumsi = 40 # 1 liter untuk 40 km
  harga_per_liter = 13000

  liter_dibutuhkan = jarak / konsumsi
  total_biaya = liter_dibutuhkan * harga_per_liter

  print("Liter dibutuhkan:", liter_dibutuhkan)
  print("Total biaya bensin;", total_biaya)
```

```python
from typing import Protocol
#4. Gaji Besih Karyawan
def gaji_besih():
  print("\n=== Gaji Bersih ===")
  gaji_pokok = float(input("Masukan gaji pokok:"))
  tunjangan = float(input("Masukan tunjangan:"))

  bpjs = 0.02
  pajak = 0.05

  potongan = (gaji_pokok * bpjs) + (gaji_pokok * pajak)
  gaji_total = gaji_pokok + tunjangan - potongan

  print("Total potongan:", potongan)
  print("Gaji bersih diterima:", gaji_total)
```

```python
#5. Penyusutan Nilai Barang (Garis Lurus)
def penyusutan():
  print("\n=== Penyusutan Nilai Barang ===")
  harga_awal = float(input("Masukan harga awal:"))
  nilai_sisa = float(input("Masukan nilai sisa:"))
  umur = int(input("Masukan umur ekonomis(tahun):"))

  penyusutan_per_tahun = (harga_awal - nilai_sisa) / umur
  nilai_setelah_2_tahun = harga_awal - (2 * penyusutan_per_tahun)

  print("Penyusutan per tahun:", penyusutan_per_tahun)
  print("Nilai setelah 2 tahun:", nilai_setelah_2_tahun)

  #Menu
  while True:
    print("\=== Menu ===")
    print("1. Spil Bill")
    print("2. Freelance")
    print("3. Estimasi BBM")
    print("4. Gaji Bersih")
    print("5. Penyusutan")
    print("0. Keluar")

    pilihan = input("Pilih menu:")

    if pilihan == "1"
      split_bill()
    elif pilihan == "2"
      freelance()
    elif pilihan == "3"
      estimasi_bbm()
    elif pilihan == "4"
      gaji_besih()
    elif pilihan == "5"
      penyusutan()
    elif pilihan == "0"
      print("Terima kasih!")
      break
    else:
      print("Pilihan tidak valid")


```
